```python
import numpy as np
import matplotlib.pyplot as plt

from nets.Activation_functions import *
border=4
x_data = np.linspace(-border,border,800)
```

```python
#testing Relu and Leacky relu 
x_data = np.linspace(-10,10,100)
llrr=LeakyReLu()

llrr.forward(x_data,0.1)
y_data=llrr.output
llrr.backward(y_data)
dy_data=llrr.updated_dinputs

# dy_data = llrr.backward(y_data)

# # # Graph
# plt.plot(x_data, y_data, x_data, dy_data)
# plt.title('leaky ReLU Activation Function & Derivative')
# plt.legend(['leaky_ReLU','derivative_leaky_ReLU'])
# plt.grid()
# plt.show()

# Setup centered axes
fig, ax = plt.subplots(figsize=(9, 5), dpi=400)
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')# Create and show plot
ax.plot(x_data,y_data, color="#307EC7", linewidth=3, label="LeackyReLU")
ax.plot(x_data,dy_data, color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="right", frameon=False)
fig.show()
```

```python
# testing tanh
llrr=tangenthyperbolic()

llrr.forward(x_data)
y_data=llrr.output
llrr.backward(y_data)
dy_data=llrr.updated_dinputs

# Setup centered axes
fig, ax = plt.subplots(figsize=(9, 5), dpi=400)
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')# Create and show plot
ax.plot(x_data,y_data, color="#307EC7", linewidth=3, label="tanh")
ax.plot(x_data,dy_data, color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="lower right", frameon=False)
fig.show()
```
